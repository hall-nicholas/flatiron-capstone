```python
import numpy as np
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
import cv2
```

```python
dataDir='../data'
trainDir='train2017'
valDir='val2017'

trainAnn=f'{dataDir}/annotations/instances_{trainDir}.json'
valAnn=f'{dataDir}/annotations/instances_{valDir}.json'

cocoTrain=COCO(trainAnn)
cocoVal=COCO(valAnn)
```

```python
catIds = cocoTrain.getCatIds(catNms=['person']);
catIds
```

```python
imgIds = cocoTrain.getImgIds(catIds=catIds)
len(imgIds)
```

```python
imgIds_2 = cocoTrain.getImgIds(imgIds=imgIds)
len(imgIds_2)
```

```python
imgIds == imgIds_2
```

```python
temp = imgIds_2[np.random.randint(0,len(imgIds_2))]
img = cocoTrain.loadImgs(temp)[0]
```

```python
fpath = '{}/{}/{}'.format(dataDir, trainDir, img['file_name'])

test_img = cv2.imread(fpath)
#cv2.imshow('random COCO photo with people', test_img)

# clears image on key press
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
